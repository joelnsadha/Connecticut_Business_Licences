```python
import pandas as pd
import psycopg2
import requests
import json
```

```python
# Get data
response = requests.get("https://data.ct.gov/resource/ngch-56tr.json")
```

```python
response_json = response.json()
```

```python
# Save data in .txt file in order to not hit the endpoint repeatedly
with open("data.txt", 'w') as file:
    json.dump(response_json, file)
```

```python
# Access text file
with open("data.txt", 'r') as my_file:
    data = json.load(my_file)
```

```python
# Store businesses in a list
businesses = []
for business in data:
    business_info = [
        business.get('credentialid'),
        business.get('name'),
        business.get('fullcredentialcode'),
        business.get('credentialtype'),
        business.get('credentialnumber'),
        business.get('credential'),
        business.get('status'),
        business.get('statusreason'),
        business.get('active'),
        business.get('issuedate'),
        business.get('effectivedate'),
        business.get('address'),
        business.get('city'),
        business.get('state'),
        business.get('zip'),
        business.get('recordrefreshedon')
    ]

    businesses.append(business_info)
```
